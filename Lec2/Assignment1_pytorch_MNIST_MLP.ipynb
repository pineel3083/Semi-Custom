### 1. Data Preparation

**Data Set**  
$$X_{train} \in \mathcal{R}^{50000 \times 784}, Y_{train} \in \mathcal{Z}^{50000}$$  
$$X_{val} \in \mathcal{R}^{10000 \times 784}, Y_{val} \in \mathcal{Z}^{10000}$$  
$$X_{test} \in \mathcal{R}^{10000 \times 784}, Y_{test} \in \mathcal{Z}^{10000}$$

```python
import torch
from torchvision import datasets, transforms
import torch.nn as nn


batch_size = 128
train_dataset = datasets.MNIST('./data', train=True, download=True,
                   transform=transforms.Compose([
                       transforms.ToTensor(),
                       transforms.Normalize((0.1307,), (0.3081,))
                   ]))
test_dataset =  datasets.MNIST('./data', train=False, download=True,
                   transform=transforms.Compose([
                       transforms.ToTensor(),
                       transforms.Normalize((0.1307,), (0.3081,))
                   ]))
train_dataset, val_dataset = torch.utils.data.random_split(train_dataset, [50000, 10000])
print(len(train_dataset), len(val_dataset), len(test_dataset))

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=128, shuffle=True)
val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=128, shuffle=False)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=128, shuffle=False)
```

### 2. Model Architecture

```python
class LinearModel(nn.Module):
    def __init__(self):
        super(LinearModel, self).__init__()
        self.linear = nn.Linear(in_features=784, out_features=10, bias=True)

    def forward(self, x):
        x = self.linear(x)
        return x

class MLPModel(nn.Module):
    def __init__(self, hid_unit, hid_layer):
        super(MLPModel, self).__init__()
        self.linear1 = nn.Linear(in_features=784, out_features=hid_unit)
        self.linear2 = nn.Linear(in_features=hid_unit, out_features=10)
        self.linear3 = nn.Linear(in_features=hid_unit, out_features=hid_unit)
        self.relu = nn.ReLU()
        self.hidden_layer = hid_layer

    def forward(self, x):
        x = self.relu(self.linear1(x))
        for i in range(self.hidden_layer-1):
          x = self.relu(self.linear3(x))
        x = self.linear2(x)
        return x


class MLPModel_ModuleList(nn.Module):
    def __init__(self, hid_unit, hid_layer):
        super(MLPModel_ModuleList, self).__init__()

        # 1. 입력층 -> 첫 번째 은닉층
        self.input_layer = nn.Linear(in_features=784, out_features=hid_unit)
        self.relu = nn.ReLU()

        # 2. 중간 은닉층들 (hid_layer - 1 개만큼 생성)
        # 💡 일반 list가 아닌 nn.ModuleList에 담아야 파이토치가 파라미터를 인식함!
        self.hidden_layers = nn.ModuleList()
        for _ in range(hid_layer - 1):
            self.hidden_layers.append(nn.Linear(in_features=hid_unit, out_features=hid_unit))

        # 3. 마지막 은닉층 -> 출력층 (클래스 개수인 10개로 출력)
        self.output_layer = nn.Linear(in_features=hid_unit, out_features=10)

    def forward(self, x):
        # 1. 입력층 통과
        x = self.relu(self.input_layer(x))

        # 2. ModuleList에 담긴 은닉층들을 순서대로 꺼내서 통과 (가중치 공유 X, 매번 다른 레이어!)
        for layer in self.hidden_layers:
            x = self.relu(layer(x))

        # 3. 출력층 통과 (마지막엔 보통 활성화 함수를 안 씌움)
        x = self.output_layer(x)
        return x
```

### 3. Cost Function Define (Loss Function Define)  

```python
cls_loss = nn.CrossEntropyLoss()
```

### 4. Train & Evaluation

```python
import torch.optim as optim
from sklearn.metrics import accuracy_score

# ====== Construct Model ====== #
model = MLPModel_ModuleList(128, 2)
# model = MLPModel(128, 10)
print('Number of {} parameters'.format(sum(p.numel() for p in model.parameters() if p.requires_grad)))

# ===== Construct Optimizer ====== #
lr = 0.005
optimizer = optim.SGD(model.parameters(), lr=lr)

list_epoch = []
list_train_loss = []
list_val_loss = []
list_acc = []
list_acc_epoch = []

epoch = 30
for i in range(epoch):

    # ====== Train ====== #
    train_loss = 0
    model.train()


    first_batch = True

    for input_X, true_y in train_loader:
        optimizer.zero_grad() # [21.01.05 오류 수정] 매 Epoch 마다 .zero_grad()가 실행되는 것을 매 iteration 마다 실행되도록 수정했습니다.

        input_X = input_X.squeeze()
        input_X = input_X.view(-1, 784)
        pred_y = model(input_X)

        loss = cls_loss(pred_y.squeeze(), true_y)
        loss.backward()

        # ========================================================== #
        # 🔍 가중치와 기울기 붕괴 관찰 코드 추가 (첫 번째 배치만 출력)
        # ========================================================== #
        if first_batch:
            # linear3의 가중치(weight)와 기울기(grad)의 평균적인 크기(norm) 추출

            w_norm = model.input_layer.weight.norm().item()
            g_norm = model.input_layer.weight.grad.norm().item() if model.input_layer.weight.grad is not None else 0.0
            print(f"[Epoch {i}] input_layer - Weight Norm: {w_norm:.4f} | Gradient Norm: {g_norm:.4f}")


            # w_norm = model.linear1.weight.norm().item()
            # g_norm = model.linear1.weight.grad.norm().item() if model.linear1.weight.grad is not None else 0.0
            # print(f"[Epoch {i}] linear - Weight Norm: {w_norm:.4f} | Gradient Norm: {g_norm:.4f}")


            first_batch = False
        # ========================================================== #

        optimizer.step()
        train_loss += loss.detach().numpy()
    train_loss = train_loss / len(train_loader)
    list_train_loss.append(train_loss)
    list_epoch.append(i)


    # ====== Validation ====== #
    val_loss = 0
    model.eval()
    # optimizer.zero_grad() [21.01.05 코드 클린업] 아래 torch.no_grad()가 호출되고 Validation 과정에서는 Optimizer를 사용하지 않으므로 굳이 호출될 필요가 없습니다.

    with torch.no_grad(): # [21.01.05 코드 클린업] Validation 과정에서 그라디언트가 활용되지 않도록 torch.no_grad() statement 안에서 코드를 실행합니다.
        for input_X, true_y in val_loader:
            input_X = input_X.squeeze()
            input_X = input_X.view(-1, 784)
            pred_y = model(input_X)

            loss = cls_loss(pred_y.squeeze(), true_y)
            val_loss += loss.detach().numpy()
        val_loss = val_loss / len(val_loader)
        list_val_loss.append(val_loss)


    # ====== Evaluation ======= #
    correct = 0
    model.eval()
    # optimizer.zero_grad() [21.01.05 코드 클린업] 아래 torch.no_grad()가 호출되고 Evaluation 과정에서는 Optimizer를 사용하지 않으므로 굳이 호출될 필요가 없습니다.

    with torch.no_grad(): # [21.01.05 코드 클린업] Evaluation 과정에서 그라디언트가 활용되지 않도록 torch.no_grad() statement 안에서 코드를 실행합니다.
        for input_X, true_y in test_loader:
            input_X = input_X.squeeze()
            input_X = input_X.view(-1, 784)
            pred_y = model(input_X).max(1, keepdim=True)[1].squeeze()
            correct += pred_y.eq(true_y).sum()

        acc = correct.numpy() / len(test_loader.dataset)
        list_acc.append(acc)
        list_acc_epoch.append(i)

    print('Epoch: {}, Train Loss: {}, Val Loss: {}, Test Acc: {}%'.format(i, train_loss, val_loss, acc*100))
```

```python
import torch
import torch.optim as optim
import numpy as np
import random
import matplotlib.pyplot as plt

# ========================================== #
# 1. Seed 고정 함수 (재현성을 위해)
# ========================================== #
def set_seed(seed=123):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed) # multi-GPU
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
    np.random.seed(seed)
    random.seed(seed)

set_seed(123) # 시드 고정!

# 각 hid_layer별 최종 Test 정확도를 저장할 딕셔너리
final_results = {}

# ========================================== #
# 2. hid_layer 1부터 5까지 반복 실험
# ========================================== #
for hid_layer in range(1, 6):
    print(f"\n{'='*50}")
    print(f"🚀 모델 학습 시작: hid_layer = {hid_layer}")
    print(f"{'='*50}")

    # ====== Construct Model ====== #
    model = MLPModel_ModuleList(128, hid_layer)
    print('Number of parameters: {}'.format(sum(p.numel() for p in model.parameters() if p.requires_grad)))

    # ===== Construct Optimizer ====== #
    lr = 0.005
    optimizer = optim.SGD(model.parameters(), lr=lr)

    # 💡 새로운 모델을 학습할 때마다 리스트를 초기화해줍니다.
    list_epoch = []
    list_train_loss = []
    list_val_loss = []
    list_acc = []
    list_acc_epoch = []

    epoch = 30
    for i in range(epoch):

        # ====== Train ====== #
        train_loss = 0
        model.train()

        first_batch = True

        for input_X, true_y in train_loader:
            optimizer.zero_grad()

            input_X = input_X.squeeze()
            input_X = input_X.view(-1, 784)
            pred_y = model(input_X)

            loss = cls_loss(pred_y.squeeze(), true_y)
            loss.backward()

            # 가중치와 기울기 붕괴 관찰 코드
            if first_batch:
                w_norm = model.input_layer.weight.norm().item()
                g_norm = model.input_layer.weight.grad.norm().item() if model.input_layer.weight.grad is not None else 0.0
                # 출력이 너무 많아질 수 있으니 10 에폭 단위로만 출력하도록 살짝 조정했습니다.
                if i % 10 == 0:
                    print(f"[Epoch {i}] input_layer - Weight Norm: {w_norm:.4f} | Gradient Norm: {g_norm:.4f}")
                first_batch = False

            optimizer.step()
            train_loss += loss.detach().numpy()

        train_loss = train_loss / len(train_loader)
        list_train_loss.append(train_loss)
        list_epoch.append(i)

        # ====== Validation ====== #
        val_loss = 0
        model.eval()

        with torch.no_grad():
            for input_X, true_y in val_loader:
                input_X = input_X.squeeze()
                input_X = input_X.view(-1, 784)
                pred_y = model(input_X)

                loss = cls_loss(pred_y.squeeze(), true_y)
                val_loss += loss.detach().numpy()

            val_loss = val_loss / len(val_loader)
            list_val_loss.append(val_loss)

        # ====== Evaluation ======= #
        correct = 0
        model.eval()

        with torch.no_grad():
            for input_X, true_y in test_loader:
                input_X = input_X.squeeze()
                input_X = input_X.view(-1, 784)
                pred_y = model(input_X).max(1, keepdim=True)[1].squeeze()
                correct += pred_y.eq(true_y).sum()

            acc = correct.numpy() / len(test_loader.dataset)
            list_acc.append(acc)
            list_acc_epoch.append(i)

        # 출력이 너무 길어지는 것을 방지하기 위해 10 에폭마다 출력
        if (i+1) % 10 == 0 or i == 0:
            print('Epoch: {:2d}, Train Loss: {:.4f}, Val Loss: {:.4f}, Test Acc: {:.2f}%'.format(i+1, train_loss, val_loss, acc*100))

    # 해당 hid_layer의 최종 Test 정확도 저장
    final_test_acc = list_acc[-1] * 100
    final_results[hid_layer] = final_test_acc
    print(f"🎯 [결과] hid_layer {hid_layer}의 최종 Test 정확도: {final_test_acc:.2f}%\n")


# ========================================== #
# 3. 모든 실험 결과 한눈에 보기
# ========================================== #
print("📊 [최종 요약] hid_layer 깊이에 따른 Test Accuracy 변화")
for hl, acc in final_results.items():
    print(f" - hid_layer {hl}: {acc:.2f}%")

# 간단한 그래프로 확인하기
layers = list(final_results.keys())
accuracies = list(final_results.values())

plt.figure(figsize=(8, 5))
plt.plot(layers, accuracies, marker='o', linestyle='-', color='b')
plt.title('Test Accuracy vs Hidden Layers (MLPModel_ModuleList)')
plt.xlabel('Number of Hidden Layers (hid_layer)')
plt.ylabel('Test Accuracy (%)')
plt.xticks(layers)
plt.grid(True)
plt.show()
```
